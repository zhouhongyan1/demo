```python
import pandas as pd
import numpy as np
```

```python
names=['user_id','user_chat_duration','is_retained','is_new','send_coin_num','receive_coin_num','chat_duration','link_mic_duration'
        ,'msg_cnt','avg_send_coin_num','avg_receive_coin_num','avg_chat_duration','avg_link_mic_duration','avg_msg_cnt','enter_uv_all'
        ,'max_users','femal_enter_uv','room_coin_num','on_mic_uv','room_msg_cnt']
data =  pd.read_csv("/Users/zhouhongyan/Documents/python/reco.txt",sep=',', encoding='utf8',header = None,names=names)

```

```python
data.head()
```

--用户信息
<br/>user_chat_duration:聊天室时长</br>
<br/>duration_level:聊天室时长区间</br>
<br/>is_retained:是否留存</br>
<br/>is_new:是否新用户</br>

--房间人指标(用户最近一周的数据和)
<br/>avg_send_coin_num:人均送电丸币</br>
<br/>avg_receive_coin_num:人均收电丸币</br>
<br/>avg_chat_duration:人均聊天室时长</br>
<br/>avg_link_mic_duration:人均连麦时长</br>
<br/>avg_msg_cnt:人均消息量</br>

--房间指标(一天的数据汇总)
<br/>enter_uv_all:进入总人数</br>
<br/>max_users:人数峰值</br>
<br/>femal_enter_uv:女性人数</br>
<br/>room_coin_num:消费电丸数</br>
<br/>on_mic_uv:上麦人数</br>
<br/>room_msg_cnt:消息量</br>

```python
target_name=['人均送电丸币','人均收电丸币','人均聊天室时长','人均连麦时长','人均消息量','进入总人数','人数峰值','女性人数'
                          ,'消费电丸数','上麦人数','消息量']
```

```python
data['duration_level']=np.where((data['user_chat_duration']>0) & (data['user_chat_duration']<=10),'0-10分钟','0')
data['duration_level']=np.where((data['user_chat_duration']>10) & (data['user_chat_duration']<=30),'10-30分钟',data['duration_level'])
data['duration_level']=np.where(data['user_chat_duration']>30,'30分钟以上',data['duration_level'])
```

```python
data.info()
```

```python
#进入聊天室用户数
data.drop_duplicates('user_id')['user_id'].count()
```

```python
#新用户数
data.loc[data['is_new']==1,:].drop_duplicates('user_id')['user_id'].count()
```

```python
data.describe()
```

```python
#计算相关性矩阵
corr_df=data.corr()
corr_df
```

```python
#提取各特征与用户聊天室时长（user_chat_duration）的相关系数，并降序排列
print('各指标与用户聊天室时长（user_chat_duration）的相关系数：')
corr_duration=corr_df['user_chat_duration']
corr_duration=corr_duration[['avg_send_coin_num','avg_receive_coin_num','avg_chat_duration','avg_link_mic_duration'
                                          ,'avg_msg_cnt','enter_uv_all','max_users','femal_enter_uv','room_coin_num'
                                         ,'on_mic_uv','room_msg_cnt']]
corr_duration=pd.DataFrame(corr_duration)
corr_duration['指标含义']=target_name
corr_duration[['指标含义','user_chat_duration']].round(4).sort_values('user_chat_duration',ascending=False)
```

```python
print('不同聊天室时长用户各指标均值：')
duration_mean=data.groupby('duration_level')['avg_send_coin_num','avg_receive_coin_num','avg_chat_duration','avg_link_mic_duration'
                                          ,'avg_msg_cnt','enter_uv_all','max_users','femal_enter_uv','room_coin_num'
                                         ,'on_mic_uv','room_msg_cnt'].mean().T
duration_mean['指标含义']=target_name
duration_mean[['指标含义','0-10分钟','10-30分钟','30分钟以上']].round(2)
```

```python
#提取各特征与留存情况（is_retained）的相关系数，并降序排列
print('各指标与留存情况（is_retained）的相关系数：')
corr_retained=corr_df['is_retained'].sort_values(ascending=False)
corr_retained=corr_retained[['avg_send_coin_num','avg_receive_coin_num','avg_chat_duration','avg_link_mic_duration'
                                          ,'avg_msg_cnt','enter_uv_all','max_users','femal_enter_uv','room_coin_num'
                                         ,'on_mic_uv','room_msg_cnt']]
corr_retained=pd.DataFrame(corr_retained)
corr_retained['指标含义']=target_name
corr_retained[['指标含义','is_retained']].round(4).sort_values('is_retained',ascending=False)
```

```python
print('留存_流失用户各指标均值：')
retained_mean=data.groupby('is_retained')['avg_send_coin_num','avg_receive_coin_num','avg_chat_duration','avg_link_mic_duration'
                                          ,'avg_msg_cnt','enter_uv_all','max_users','femal_enter_uv','room_coin_num'
                                         ,'on_mic_uv','room_msg_cnt'].mean().T
retained_mean['指标含义']=target_name
retained_mean[['指标含义',1,0]].round(2)
```
